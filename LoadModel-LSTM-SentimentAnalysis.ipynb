```python
from keras.models import load_model
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
# returns a compiled model
# identical to the previous one
model = load_model('Waris_LSTM-SentimentAnalysis.h5')
```

```python
tokenizer = Tokenizer(num_words = 1000, oov_token="<OOV>")
sample_string = "bagus banget aplikasinya"
tokenizer.fit_on_texts(sample_string)
sample = tokenizer.texts_to_sequences(sample_string)
padded_sample = pad_sequences(sample, padding='post')
y_pred = model.predict(padded_sample.T)[0][0]
print("Padded sample", padded_sample.T)
print("Probabilty of a person sentiment :",y_pred*100,"%")
# "Bad" if np.argmax(y_pred) == 0 else "Good"
```
